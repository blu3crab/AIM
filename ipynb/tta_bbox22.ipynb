```python
# Model B
# Define the predictions object with dummy data for testing
predictions = [
    {
        "word": "hello",
        "bounding_box": [10, 20, 50, 70]
    },
    {
        "word": "world",
        "bounding_box": [60, 80, 100, 120]
    }
]

def preds_to_json(base_filename, predictions):
    import json
    json_filename = base_filename + ".json"
    with open(json_filename, "w") as json_file:
        json.dump(predictions, json_file, indent=4)

def json_to_preds(base_filename, predictions):
    import json
    json_filename = base_filename + ".json"
    with open(json_filename, "r") as json_file:
        return json.load(json_file)

# Test code
base_filename = "test_preds"
preds_to_json(base_filename, predictions)
read_predictions = json_to_preds(base_filename, predictions)

# Compare the read predictions with the original
are_equal = read_predictions == predictions

# Summarize the results
print("Predictions are equal:", are_equal)

# Remove the test files
import os
os.remove(base_filename + ".json")
```

```python
# Model B
import numpy as np
import json

# Define the predictions object with actual data
predictions = [
    ('of', np.array([[1812.8906 ,  434.29688],
       [1972.2656 ,  434.29688],
       [1972.2656 ,  525.9375 ],
       [1812.8906 ,  525.9375 ]], dtype=np.float32)),
    ('america', np.array([[2039.0361 ,  440.46344],
       [2700.8264 ,  430.73114],
       [2702.342  ,  533.7752 ],
       [2040.5515 ,  543.5075 ]], dtype=np.float32)),
    ('states', np.array([[1221.82   ,  484.05478],
       [1766.906  ,  442.125  ],
       [1774.8046 ,  544.8047 ],
       [1229.7185 ,  586.73444]], dtype=np.float32)),
    ('united', np.array([[ 640.5678 ,  508.20175],
       [1172.9049 ,  472.71265],
       [1180.0449 ,  579.8147 ],
       [2621.7188, 4004.2969]], dtype=np.float32))
]

def preds_to_json(base_filename, predictions):
    # Convert numpy arrays to lists as JSON can't handle numpy arrays
    json_friendly_preds = []
    for key, val in predictions:
      json_friendly_preds.append({
          'key': key,
          'val': val.tolist()
      })

    json_filename = base_filename + ".json"
    with open(json_filename, "w") as json_file:
        json.dump(json_friendly_preds, json_file, indent=4)

def json_to_preds(base_filename, predictions):
    json_filename = base_filename + ".json"
    with open(json_filename, "r") as json_file:
        json_data = json.load(json_file)

    # Convert JSON lists back to numpy arrays
    converted_preds = []
    for item in json_data:
      converted_preds.append((
          item['key'],
          np.array(item['val'])
      ))
    return converted_preds

# Test code
base_filename = "test_preds"
preds_to_json(base_filename, predictions)
read_predictions = json_to_preds(base_filename, predictions)

# Compare the read predictions with the original
# are_equal = read_predictions == predictions
are_equal = np.array_equal(read_predictions,predictions)
print(read_predictions)
print(predictions)
# Summarize the results
print("Predictions are equal:", are_equal)

# Remove the test files
import os
os.remove(base_filename + ".json")
```
